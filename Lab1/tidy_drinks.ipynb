```python
import pandas as pd
import datetime
from os import listdir
from os.path import isfile, join
import glob
import re
import shutil
df = pd.read_csv("./Original Data/drinks.csv")
df=df.drop(columns=['total_litres_of_pure_alcohol'])
df
```

```python
# Format to tidy data
formatted_df = pd.melt(df,
                       ["country"],
                       var_name="servings",
                       value_name="freq")
formatted_df = formatted_df.sort_values(by=["country"])
formatted_df.head(10)
```

```python
formatted_df.to_csv('tidy_drinks.csv', index=False)
# Source path 
source = 'tidy_drinks.csv'
  
# Destination path 
destination ='Analysis Data'
  
# Move the content of 
# source to destination 
dest = shutil.move(source, destination)
```
